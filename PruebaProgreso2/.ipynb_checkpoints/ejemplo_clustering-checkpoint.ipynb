# Ejemplo completo de clustering

## Importar librerías
A mí me gusta como primer paso importar juntas todas las librerías que se vayan a usar. Python permite importarlas sólo cuando hace falta, pero muchas veces eso supone añadir cómputo extra a las celdas que cuando se ejecutan varias veces puede ralentizar el proceso.

```python
import numpy as np
import pandas as pd
from sklearn.cluster import DBSCAN, KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns
```

## Cargar los datos
Ahora vamos a cargar los datos. Estos consisten en un conjunto de características físicas de diferentes modelos de coches (como su "circularidad", "elongamiento" o "compactitud") y una etiqueta indicando el tipo de vehículo en cuestión. Lo ideal para poder trabajar sería entender por completo en qué consisten los datos, pero en el repositorio no nos dan esa información. Si investigamos por separado cada característica, es posible obtener información de qué significan, pero en este caso vamos a optar por hacer un análisis ciego al significado de los campos. Son menos potentes, pero pueden resultar útiles a su vez para entender qué nos dicen los datos.

A parte de cargar los datos, vamos a eliminar la columna que representa la clase de coche, ya que para una labor de clustering no nos será particularmente útil.

```python
# carga del fichero
ruta =  "https://raw.githubusercontent.com/milaan9/Clustering-Datasets/refs/heads/master/01.%20UCI/vehicle.csv"
data = pd.read_csv(ruta)
# eliminamos la clase de vehículo, ya que no será necesaria para una labor de clustering
data = data.drop(columns = ["Class"])
data.head()
```

## Análisis exploratorio
Ahora vamos a echar un ojo en general a los datos. En este caso hay algo más de 800 filas y como humanos, revisarlas una a una no solo sería costoso, sino poco útil, así que necesitamos echar un ojo general. Usaremos 3 herramientas para ello, por un lado el método .describe() que proporciona pandas para sus DataFrames que calcula automáticamente varios descriptores estadísticos, como la media, la desviación típica o la mediana, entre otros. La segunda herramienta será graficar los histogramas que nos permitan visualizar la distribución de los datos. La última herramienta será una matriz de correlación que nos indica cuán relacionadas entre sí están las diferentes variables.

```python
# resúmenes estadísticos de los datos
data.describe()
```

Echando un ojo a los estadísticos descriptivos, podemos apreciar que hay bastantes diferencias entre los valores entre los que se mueve cada característica. También vemos una alta variedad de desviaciones típicas, lo que indica que hay algunos grupos de datos muy dispersos y otros menos.

```python
# visualización de la distribución de cada variable
plt.figure(figsize=(9, 12))
for i, column in enumerate(data.columns):
    plt.subplot(6, 3, i+1)
    sns.histplot(data[column], kde=True)
    plt.xlabel("")
    plt.ylabel("")
    plt.title(f'{column}')
plt.tight_layout()
plt.show()
```

Las distribuciones muestran una gran variabilidad y algunos casos curiosos. Por ejemplo, se puede apreciar que varias características tienen una distribución normal (campana de Gauss), pero son relativamente pocas y la mayoría tienen distribuciones menos habituales. Entre esas distribuciones menos habituales, algunas puede apreciarse que están desplazadas a la derecha o izquierda, otras tienen una parte central especialmente amplia y otras parecen la suma de 2 distribuciones normales. Esta gran variabilidad puede ser un indicativo de que las características son bastante independientes unas de otras.

```python
# cálculo de correlaciones
corrs = data.corr()
# graficar todas las correlaciones
sns.heatmap(corrs)
plt.show()
# destacar sólo aquellas características muy correlacionadas
sns.heatmap(corrs > .95)
plt.show()
# eliminar las que más se parecen entre sí
data = data.drop(columns = ["'SCATTER RATIO'", "CIRCULARITY", "'PR.AXIS RECTANGULARITY'"])
```

Como puede apreciarse hay muy pocas características muy correlacionadas entre sí, lo que indica que casi todas aportan información útil. En este caso hemos decidido eliminar 3 porque aportan una información muy similar a otras características y pueden ser un poco redundantes e innecesarias.

## Normalización
El siguiente paso, antes de ponernos a analizar los clusters como tal, es normalizar los datos. Habitualmente, es recomendable normalizarlos siempre que se trabaja con algoritmos de aprendizaje automático, aunque no todos lo necesitan. En el clustering es particularmente importante, porque las características con una escala mayor van a ser más relevantes que las que tienen una escala más pequeña. Esto también puede usarse a favor a veces si se quiere priorizar que el clustering se centre en algunos campos específicos, aunque no se quiera olvidar al resto.

```python
# normalización los datos
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
scaled_data = pd.DataFrame(scaled_data, columns = data.columns)
```

## Selección de número de clusters para K-Means
El siguiente código muestra la evaluación del índice de Silhouette para diferentes números de clusters utilizando el algoritmo K-Means. El mínimo de clusters a probar es 2, porque no tiene mucho sentido empezar con solo 1 y el máximo es 11 porque ya es un número relativamente alto teniendo en cuenta la cantidad de datos. Al inicializar K-Means, se utilizan 2 opciones a tener en cuenta. Por un lado está **random_state** que se usa para asegurarse que los resultados siempre sean los mismos cada ejecución, ya que en K-Means hay aleatoriedad envuelta a la hora de seleccionar los centroides. La opción **n_init** sirve para que pruebe diferentes centroides iniciales antes de quedarse con los definitivos, y en este caso lo estamos usando también para que el efecto de la aleatoriedad se reduzca, ya que, en este caso específico, si no se usara ese parámetro, al cambiar el random_state los resultados pueden cambiar mucho. La conclusión que sacamos del análisis de Silhouette es que el mejor resultado lo da con 2 clusters, pero con 4 también da un resultado a considerar, mientras que el resto es bastante bajo.

```python
# evaluación Silhouette para KMeans
silhouette_scores = []
for n_clusters in range(2, 11):
    kmeans = KMeans(n_clusters=n_clusters, n_init = 10, random_state=42)
    cluster_labels = kmeans.fit_predict(scaled_data)
    silhouette_avg = silhouette_score(scaled_data, cluster_labels)
    silhouette_scores.append(silhouette_avg)

plt.figure(figsize=(10, 6))
plt.plot(range(2, 11), silhouette_scores, marker='o', linestyle='--', color='r')
plt.title('Análisis de Silhouette')
plt.xlabel('Número de clusters')
plt.ylabel('Puntuación Silhouette')
plt.show()
```

## Clustering
Ahora vamos a pasar al clustering propiamente dicho. Como puede apreciarse, DBSCAN está ya inicializado con unos parámetros bastante específicos. La forma de obtenerlos ha sido repetir la ejecución varias veces hasta que los resultados tenían "buena pinta". Esa "buena pinta" significa que al menos haya 2 clusters y no demasiados valores atípicos. Aún así, como puede verse en este ejemplo hay bastantes. Una posible interpretación de esto, es que haya varios tipos de coches muy similares y otros tan lejos de la norma que es complicado asociarlos a un grupo.

El resultado de K-Means muestra una separación en 2 clusters que tiene bastante buena pinta a simple vista y que viene además verificada por el índice de Silhouette.

Es importante tener en cuenta que esta es una visualización 2D obtenida tras aplicar el algoritmo PCA y que no nos muestra la distribución exacta de los datos y que necesitaríamos más de 10 dimensiones para mostrarlos correctamente, lo cual es imposible de visualizar para seres humanos (y sospecho que también para cualquier otro bicho conocido).

```python
seed = 42
# K-means clustering
kmeans = KMeans(n_clusters=2, n_init=10, random_state = 42)
kmeans_labels = kmeans.fit_predict(scaled_data)
data["kmeans_clusters"] = kmeans_labels

# DBSCAN clustering
dbscan = DBSCAN(eps=1.63, min_samples=10)
dbscan_labels = dbscan.fit_predict(scaled_data)
data["dbscan_clusters"] = dbscan_labels


# preparación de la visualización
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# transformación datos a 2D para poder visualizarlos
pca = PCA(2)
data_2d = pca.fit_transform(scaled_data)

# visualización K-means
ax1.scatter(data_2d[:, 0], data_2d[:, 1], c=kmeans_labels, cmap='Paired')
ax1.set_title('K-means Clustering')

# visualización DBSCAN
outliers_i = (dbscan_labels == -1)
ax2.scatter(data_2d[~outliers_i, 0], data_2d[~outliers_i, 1], c=dbscan_labels[~outliers_i], cmap='Paired')
# se muestran por separado los outliers, para no usar el mismo código de colores que con el resto,
# lo que ayuda a identificarlos en la figura 
ax2.scatter(data_2d[outliers_i, 0], data_2d[outliers_i, 1], color = "black")
ax2.set_title('DBSCAN Clustering')

plt.show()
```

## Análisis de clusters
Por último, es interesante estudiar qué caracteriza cada cluster y para ellos vamos a volver con los estadísticos descriptivos. Este tipo de resúmenes es muy apropiado para clústers generados por K-Means por su forma más redonda, aunque para otro tipo de clusters puede no ser tan útil. En ese caso, por la forma que aparentan tener en la proyección 2D de PCA, parece que también son apropiados para DBSCAN, a excepción de los outliers, que no tiene sentido interpretarlos como un cluster en sí mismo. La interpretación de que representa cada cluster lo dejamos para vosotras.

```python
# cálculo de valores estadísticos descriptivos
kmeans_stats = data.groupby('kmeans_clusters').agg(['mean', 'median', 'std'])
dbscan_stats = data.groupby('dbscan_clusters').agg(['mean', 'median', 'std'])
```
