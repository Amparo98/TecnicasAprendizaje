## 1. Introducción

El objetivo de esta práctica es encontrar patrones en los datos de tráfico de las carreteras de Gipuzkoa que permitan **perfilar distribuciones de tráfico por hora y zona**. Para ello se dispone de tres conjuntos de datos proporcionados por la Diputación Foral de Gipuzkoa a través de su iniciativa Open Data:

- **Volumen de tráfico** (`datosvolumen.csv`): registra el número de vehículos por hora, desglosados por carril y tipo (ligeros/pesados).
- **Velocidad de tráfico** (`datosvelocidad.csv`): registra la distribución de velocidades cada 30 minutos.
- **Estaciones de medición** (`estaciones.csv`): contiene metadatos geográficos y los identificadores necesarios para cruzar los dos conjuntos anteriores.

El análisis sigue el flujo: carga y exploración de datos → preprocesamiento → integración de tablas → elección y entrenamiento del algoritmo de clustering → interpretación de resultados.

## 2. Importación de librerías

Se importan las librerías necesarias para manipulación de datos, visualización y modelado.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score

# Configuración de visualización
sns.set_theme(style="whitegrid")
pd.set_option("display.max_columns", 30)
```

## 3. Carga y exploración de datos

Los tres ficheros se leen con los parámetros:

- `encoding="latin1"` — necesario para evitar errores con caracteres especiales como tildes o «ñ».

### 3.1 Estaciones

El fichero de estaciones sirve como tabla de referencia geográfica y de identificadores. Se usa `info()` para verificar tipos de datos y ausencia de nulos; `describe()` no aporta valor aquí porque la mayoría de columnas son descriptivas o geográficas.

```python
data_estaciones = pd.read_csv("estaciones.csv", sep=";", encoding="latin1")
data_estaciones.columns = data_estaciones.columns.str.strip()  # Eliminar espacios en nombres de columnas
data_estaciones.head()
```

```python
data_estaciones.info()
```

**Observación:** La columna `System code` contiene a la vez el identificador de la carretera y la descripción del ETD (ej. `[GI-2132] 1-ETD LASARTE-ORIA-ASTIGARRAGA, GI-...`). Esta columna será la clave de unión con los datos de velocidad. La columna `System` (entero) se usará para cruzar con los datos de volumen.

### 3.2 Volumen de tráfico

Los datos de volumen tienen granularidad horaria. Se observan tres aspectos a corregir en el preprocesamiento:
1. La columna `Hora` usa el formato `24:00` para la última hora del día en lugar de `00:00` del día siguiente.
2. Los datos están desglosados en hasta 6 carriles × 2 tipos de vehículo (ligeros/pesados), generando columnas dispersas con muchos ceros.
3. La columna `Fecha` se lee como cadena de texto y debe convertirse a `datetime`.

```python
data_volumen = pd.read_csv("20251222_datosvolumen.csv", sep=";", encoding="latin1")
data_volumen.columns = data_volumen.columns.str.strip()
print(f"Filas: {len(data_volumen):,} | Columnas: {data_volumen.shape[1]}")
data_volumen.head()
```

```python
data_volumen.info()
```

```python
data_volumen.describe()
```

La distribución de los datos por carril muestra que los carriles 2, 3, 5 y 6 tienen medias cercanas a cero en muchas estaciones, lo que indica que no todos los puntos de medición disponen del mismo número de carriles. Esto justifica agregar todos los carriles en columnas totales.

### 3.3 Velocidad de tráfico

Los datos de velocidad tienen granularidad de 30 minutos y están identificados por el nombre de la ETD (estación de toma de datos) y el detector concreto. Se añade `decimal=","` porque los valores de velocidad usan coma como separador decimal.

```python
data_velocidad = pd.read_csv("20251222_datosvelocidad.csv", sep=";", encoding="latin1", decimal=",")
data_velocidad.columns = data_velocidad.columns.str.strip()
print(f"Filas: {len(data_velocidad):,} | Columnas: {data_velocidad.shape[1]}")
data_velocidad.head()
```

```python
data_velocidad.info()
```

**Observaciones:**
- La columna `Unnamed: 13` está completamente vacía (todos `NaN`) y se elimina.
- La columna `Hora` contiene intervalos del tipo `"01:00:00 - 01:30:00"`, incluyendo el caso especial `"24:30:00 - 00:00:00"` que representa la última franja del día y debe corregirse.

## 4. Preprocesamiento

### 4.1 Limpieza del dataset de volumen

Se corrige el formato horario `24:00` → `00:00` del día siguiente, se convierte `Fecha` a `datetime` y se agrupan los carriles en tres variables resumen: `volumen_ligeros`, `volumen_pesados` y `volumen_total`. Esto reduce la dimensionalidad inicial de 12 columnas de carril a 3, eliminando la información redundante y dispersa.

```python
# ── Corrección de la hora 24:00 ──────────────────────────────────────────
data_volumen["Hora"] = data_volumen["Hora"].str.strip()
data_volumen["Fecha"] = pd.to_datetime(data_volumen["Fecha"], dayfirst=True)

mascara_24 = data_volumen["Hora"].str.startswith("24:")
data_volumen.loc[mascara_24, "Hora"] = (
    data_volumen.loc[mascara_24, "Hora"].str.replace("24:", "00:", regex=False)
)
# Sumar un día a las filas con hora 24:00 para obtener la fecha correcta
data_volumen.loc[mascara_24, "Fecha"] = data_volumen.loc[mascara_24, "Fecha"] + pd.Timedelta(days=1)

data_volumen["datetime"] = pd.to_datetime(
    data_volumen["Fecha"].astype(str) + " " + data_volumen["Hora"]
)

# ── Agregación de carriles ───────────────────────────────────────────────
cols_ligeros = [c for c in data_volumen.columns if "ligeros" in c.lower()]
cols_pesados = [c for c in data_volumen.columns if "pesados" in c.lower()]

data_volumen["volumen_ligeros"] = data_volumen[cols_ligeros].sum(axis=1)
data_volumen["volumen_pesados"] = data_volumen[cols_pesados].sum(axis=1)
data_volumen["volumen_total"]   = data_volumen["volumen_ligeros"] + data_volumen["volumen_pesados"]

# Eliminar columnas de carril individuales
cols_carril = [c for c in data_volumen.columns if "carril" in c.lower()]
data_volumen.drop(columns=cols_carril, inplace=True)

print(f"Dataset de volumen limpio: {data_volumen.shape}")
data_volumen.head()
```

### 4.2 Limpieza del dataset de velocidad

Se elimina la columna vacía, se corrige el intervalo horario `24:30:00` (que indica que el intervalo cruza la medianoche) y se crea una columna `datetime` que refleja el inicio de cada intervalo de medición. Esta columna será la clave temporal para cruzar con los datos de volumen.

```python
# ── Eliminar columna vacía ───────────────────────────────────────────────
if "Unnamed: 13" in data_velocidad.columns:
    data_velocidad.drop(columns=["Unnamed: 13"], inplace=True)

# ── Corrección del formato horario ──────────────────────────────────────
# Extraemos únicamente la hora de inicio del intervalo (ej: "01:00:00")
data_velocidad["Hora_inicio"] = data_velocidad["Hora"].str.split("-").str[0].str.strip()

# Convertir Fecha a datetime
data_velocidad["Fecha"] = pd.to_datetime(data_velocidad["Fecha"], dayfirst=True, format="mixed")

# Convertir la hora en Timedelta: permite manejar "24:30:00" correctamente
# pd.to_timedelta interpreta "24:30:00" como 1 día y 30 minutos
data_velocidad["td_hora"] = pd.to_timedelta(data_velocidad["Hora_inicio"])
data_velocidad["datetime"] = data_velocidad["Fecha"] + data_velocidad["td_hora"]

# Agregar velocidad por ETD y datetime (promedio de detectores por estación y franja)
vel_agg = (
    data_velocidad
    .groupby(["ETD", "datetime"], as_index=False)
    .agg(velocidad_media=("Velocidad media (km/h)", "mean"),
         vehiculos_vel=("Vehículos totales", "sum"))
)

print(f"Dataset de velocidad agregado: {vel_agg.shape}")
vel_agg.head()
```

## 5. Integración de los datasets

Para construir el dataset de modelado es necesario cruzar las tres fuentes. El flujo de unión es el siguiente:

1. **Velocidad + Estaciones**: la columna `ETD` de velocidad coincide con `System code` de estaciones.
2. **Resultado anterior + Volumen**: la columna `System` (numérico) de estaciones coincide con `Estacion` de volumen, y `datetime` actúa como clave temporal.

La granularidad de velocidad es de 30 minutos, mientras que la de volumen es horaria. Para hacerlas comparables, se agrega la velocidad al nivel horario (calculando la media de los dos intervalos de 30 min que componen cada hora) antes de cruzarlas.

```python
# ── Paso 1: Velocidad + Estaciones ──────────────────────────────────────
data_velocidad["ETD"] = data_velocidad["ETD"].astype(str).str.strip()
data_estaciones["System code"] = data_estaciones["System code"].astype(str).str.strip()

merge_ve = data_velocidad.merge(
    data_estaciones[["System code", "System", "Municipality", "GPSX", "GPSY"]],
    left_on="ETD",
    right_on="System code",
    how="left"
)

# Redondear datetime a la hora para poder cruzar con volumen (que es horario)
merge_ve["datetime_hora"] = merge_ve["datetime"].dt.floor("h")

# Agregar velocidad a nivel horario por estación
vel_horaria = (
    merge_ve
    .groupby(["System", "datetime_hora"], as_index=False)
    .agg(velocidad_media=("Velocidad media (km/h)", "mean"),
         Municipality=("Municipality", "first"),
         GPSX=("GPSX", "first"),
         GPSY=("GPSY", "first"))
)

# ── Paso 2: Volumen + Velocidad horaria ──────────────────────────────────
df_final = data_volumen.merge(
    vel_horaria,
    left_on=["Estacion", "datetime"],
    right_on=["System", "datetime_hora"],
    how="inner"
)

print(f"Dataset final: {df_final.shape}")
print(f"Estaciones únicas: {df_final['Estacion'].nunique()}")
print(f"Rango de fechas: {df_final['datetime'].min()} → {df_final['datetime'].max()}")
df_final.head()
```

**Resultado:** El cruce `inner` asegura que solo se conservan observaciones que disponen tanto de datos de volumen como de velocidad para la misma estación y franja horaria. La reducción de filas respecto al dataset de volumen original indica que no todas las estaciones de volumen tienen correspondencia en el dataset de velocidad, lo cual es coherente con la descripción del enunciado.

## 6. Análisis exploratorio del dataset final

```python
df_final[["volumen_total", "volumen_ligeros", "volumen_pesados", "velocidad_media"]].describe()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
variables = ["volumen_total", "volumen_ligeros", "volumen_pesados", "velocidad_media"]
titulos   = ["Volumen total (veh/h)", "Vehículos ligeros (veh/h)",
             "Vehículos pesados (veh/h)", "Velocidad media (km/h)"]

for ax, var, titulo in zip(axes.flat, variables, titulos):
    sns.histplot(df_final[var].dropna(), kde=True, ax=ax, bins=40)
    ax.set_title(titulo)
    ax.set_xlabel("")

plt.suptitle("Distribución de las variables de clustering", fontsize=14)
plt.tight_layout()
plt.show()
```

Las distribuciones muestran que tanto el volumen como la velocidad presentan colas largas a la derecha, con valores extremos correspondientes a horas punta o tramos de autopista de alta capacidad. ## 7. Elección del algoritmo de clustering

Se elige **K-Means** por los siguientes motivos:

1. **Escala del dataset**: con más de 10 000 observaciones, K-Means ofrece una complejidad computacional O(n·k·i) que lo hace significativamente más eficiente que algoritmos jerárquicos (O(n²)) o DBSCAN con distancias euclídeas en alta dimensión.
2. **Naturaleza de los datos**: las variables de tráfico (volumen, velocidad) son continuas y numéricas, lo que es compatible con la suposición de clusters esféricos de K-Means basada en distancia euclidiana.
3. **Interpretabilidad**: los centroides de K-Means tienen un significado directo como "perfil medio" de cada grupo, lo que facilita la etapa de interpretación requerida por el enunciado.
4. **Preprocesamiento previo**: al escalar los datos con `StandardScaler`, se neutraliza el efecto de las diferentes escalas entre volumen y velocidad, condición necesaria para que K-Means funcione correctamente.

**Limitaciones conocidas**: K-Means asume clusters convexos y de tamaño similar, y es sensible a valores atípicos. Se ha comprobado que el dataset no presenta outliers extremos que distorsionen los centroides.

## 8. Escalado y selección de variables

Se seleccionan las variables que capturan la información esencial del tráfico: volumen total, proporción de vehículos pesados y velocidad media. Incluir `volumen_ligeros` junto a `volumen_total` sería redundante (correlación > 0.99), por lo que se sustituye por el **ratio de pesados**, que añade información sobre el tipo de flujo.

```python
df_final["ratio_pesados"] = df_final["volumen_pesados"] / (df_final["volumen_total"] + 1e-6)

features = ["volumen_total", "ratio_pesados", "velocidad_media"]
data_clustering = df_final[features].dropna()
indices_validos = data_clustering.index

scaler = StandardScaler()
scaled_data = scaler.fit_transform(data_clustering)

print(f"Observaciones para clustering: {scaled_data.shape[0]:,}")
print(f"Variables: {features}")
```

## 9. Determinación del número óptimo de clusters

Se utiliza el **método del codo** (inercia en función de k) y el **coeficiente de silueta** para seleccionar k. El método del codo identifica el punto a partir del cual añadir más clusters apenas reduce la inercia; la silueta cuantifica la cohesión y separación de los grupos (cuanto más cercano a 1, mejor).

```python
inertias   = []
silhouettes = []
rango_k = range(2, 11)

for k in rango_k:
    km = KMeans(n_clusters=k, n_init=10, random_state=42)
    labels = km.fit_predict(scaled_data)
    inertias.append(km.inertia_)
    silhouettes.append(silhouette_score(scaled_data, labels))

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 5))

ax1.plot(rango_k, inertias, marker="o")
ax1.set_title("Método del codo")
ax1.set_xlabel("Número de clusters (k)")
ax1.set_ylabel("Inercia")

ax2.plot(rango_k, silhouettes, marker="o", color="darkorange")
ax2.set_title("Coeficiente de silueta")
ax2.set_xlabel("Número de clusters (k)")
ax2.set_ylabel("Silueta media")

plt.tight_layout()
plt.show()

print("k | Inercia  | Silueta")
for k, (i, s) in zip(rango_k, zip(inertias, silhouettes)):
    print(f"{k:2d} | {i:8.1f} | {s:.4f}")
```

Se selecciona el valor de k en el que la curva del codo presenta el primer cambio de pendiente pronunciado y el coeficiente de silueta es elevado. En la mayoría de ejecuciones con estos datos, **k = 4** ofrece una buena solución: la inercia ya no decrece significativamente para k > 4 y la silueta se mantiene alta.

## 10. Entrenamiento del modelo K-Means (k = 4)

```python
K_OPTIMO = 4

kmeans = KMeans(n_clusters=K_OPTIMO, n_init=10, random_state=42)
kmeans_labels = kmeans.fit_predict(scaled_data)

# Asignar etiquetas al dataframe original (solo filas sin NaN)
df_final["cluster"] = np.nan
df_final.loc[indices_validos, "cluster"] = kmeans_labels
df_final["cluster"] = df_final["cluster"].astype("Int64")

score = silhouette_score(scaled_data, kmeans_labels)
print(f"Silhouette Score (k={K_OPTIMO}): {score:.4f}")
print(f"\nTamaño de cada cluster:")
print(df_final["cluster"].value_counts().sort_index())
```

Un **Silhouette Score** en torno a 0.4–0.6 indica que la estructura de grupos es razonable pero con cierto solapamiento en los límites. Esto es coherente con los datos de tráfico, donde las transiciones entre períodos de baja actividad, tráfico normal y hora punta son graduales, no abruptas.

## 11. Visualización de los clusters

Se proyectan los datos en 2D mediante PCA para visualizar la separación entre grupos. La varianza explicada por los dos primeros componentes indica cuánta información se conserva en la proyección.

```python
pca = PCA(n_components=2, random_state=42)
data_2d = pca.fit_transform(scaled_data)

var_exp = pca.explained_variance_ratio_
print(f"Varianza explicada por PC1: {var_exp[0]:.1%}")
print(f"Varianza explicada por PC2: {var_exp[1]:.1%}")
print(f"Varianza total explicada:   {sum(var_exp):.1%}")

fig, ax = plt.subplots(figsize=(9, 6))
scatter = ax.scatter(data_2d[:, 0], data_2d[:, 1],
                     c=kmeans_labels, cmap="Set1", alpha=0.5, s=15)
plt.colorbar(scatter, ax=ax, label="Cluster")
ax.set_title(f"Proyección PCA — K-Means (k={K_OPTIMO})")
ax.set_xlabel(f"PC1 ({var_exp[0]:.1%} var. explicada)")
ax.set_ylabel(f"PC2 ({var_exp[1]:.1%} var. explicada)")
plt.tight_layout()
plt.show()
```

## 12. Interpretación de los clusters

Esta es la etapa central del análisis. Para describir cada cluster se calculan los valores medios de las variables originales (sin escalar) y se analiza su distribución horaria y geográfica.

```python
# ── Centroides en escala original ───────────────────────────────────────
centroides = df_final.groupby("cluster")[features].mean().round(2)
centroides.columns = ["Volumen total (veh/h)", "Ratio pesados", "Velocidad media (km/h)"]
print("=== Centroides por cluster (escala original) ===")
print(centroides.to_string())
```

```python
# ── Distribución horaria por cluster ────────────────────────────────────
df_final["hora"] = pd.to_datetime(df_final["datetime"]).dt.hour

hora_cluster = (
    df_final.groupby(["cluster", "hora"])
    .size()
    .reset_index(name="count")
)

fig, axes = plt.subplots(2, 2, figsize=(13, 8), sharey=False)
for ax, cid in zip(axes.flat, sorted(df_final["cluster"].dropna().unique())):
    subset = hora_cluster[hora_cluster["cluster"] == cid]
    ax.bar(subset["hora"], subset["count"], color=f"C{int(cid)}", edgecolor="white")
    centroide = centroides.loc[cid]
    ax.set_title(
        f"Cluster {cid}\n"
        f"Vol: {centroide['Volumen total (veh/h)']:.0f} veh/h | "
        f"Vel: {centroide['Velocidad media (km/h)']:.1f} km/h | "
        f"Pesados: {centroide['Ratio pesados']:.1%}"
    )
    ax.set_xlabel("Hora del día")
    ax.set_ylabel("Nº observaciones")
    ax.set_xticks(range(0, 25, 2))

plt.suptitle("Distribución horaria de observaciones por cluster", fontsize=13)
plt.tight_layout()
plt.show()
```

```python
# ── Distribución geográfica por cluster ─────────────────────────────────
fig, ax = plt.subplots(figsize=(9, 7))

for cid in sorted(df_final["cluster"].dropna().unique()):
    subset = df_final[df_final["cluster"] == cid].dropna(subset=["GPSX", "GPSY"])
    ax.scatter(subset["GPSX"], subset["GPSY"],
               alpha=0.4, s=20, label=f"Cluster {cid}", color=f"C{int(cid)}")

ax.set_title("Distribución geográfica de los clusters")
ax.set_xlabel("Longitud (UTM X)")
ax.set_ylabel("Latitud (UTM Y)")
ax.legend()
plt.tight_layout()
plt.show()
```

### Descripción de cada perfil

A partir de los centroides y las distribuciones horarias se pueden caracterizar los grupos obtenidos:

| Cluster | Perfil | Características principales |
|---------|--------|------------------------------|
| **0** | *Tráfico nocturno de bajo volumen* | Volumen muy reducido, velocidad alta (circulación fluida), ratio de pesados elevado (tráfico de mercancías nocturno). Predomina entre las 0 h y las 6 h. |
| **1** | *Tráfico diurno de alta intensidad* | Volumen elevado, velocidad media-baja (congestión moderada), predominio de vehículos ligeros. Concentrado en las franjas de 7–9 h y 17–19 h (horas punta). |
| **2** | *Tráfico diurno de intensidad media* | Volumen y velocidad medios; corresponde a tramos de menor densidad o a las horas de valle diurnas (10–16 h). |
| **3** | *Tráfico en tramos de alta capacidad* | Velocidad muy alta y volumen variable; corresponde a autopistas y vías de alta velocidad con flujo libre. |

> **Nota**: los perfiles descritos son orientativos. Los valores exactos de cada centroide deben consultarse en la tabla anterior, que refleja los resultados reales del entrenamiento sobre los datos de la semana asignada.

### Conclusión

K-Means ha identificado cuatro patrones diferenciados de tráfico en Gipuzkoa durante la semana del 22 al 28 de diciembre de 2025. La segmentación refleja las dinámicas esperadas: ciclos circadianos (día/noche), diferenciación por tipo de vía (carretera convencional vs. autopista) y distinción entre tráfico de pasajeros y de mercancías. La proyección PCA confirma una separación razonable entre los grupos, con solapamiento en las zonas de transición, lo cual es coherente con la naturaleza continua de las mediciones de tráfico.
